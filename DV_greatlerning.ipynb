```python
# Syntax

seaborn.barplot(data=None, 
                x=None, y=None, hue=None, 
                order=None, hue_order=None,
                estimator='mean', errorbar=('ci', 95), 
                n_boot=1000, units=None, seed=None, 
                orient=None, color=None, 
                palette=None, saturation=0.75, width=0.8, 
                errcolor='.26', 
                errwidth=None, capsize=None, dodge=True, 
                ci='deprecated', ax=None, **kwargs)
```

| Parameter | Description |
| --- | --- |
| data | **DataFrame** DataFrame Jahan sy x,y column lainy hen |
| x | **Variable** Wo variable jisy x axis par rakhna hai |
| y |**Variable** Wo variable jisy y axis par rakhna hai |
| hue |**Variable** Wo variable Jisy Colors ky hisab sy show karwaya jay |
| order |**List of String** Categorical variables ki tarteeb to change karny ka liya |
| hue_order | **List of String** hue ka Categorical variables ki tarteeb to change karny ka liya |
| estimator | Statistical function to estimate within each categorical bin, can be None to use a count. |
|errorbar| If True, draw a data-dependent error bar to indicate standard error, confidence interval, or other estimate of the central tendency of the plot. |
| ci | **int (0-100)** 0 for no confidence and 100 for full confidence |
| n_boot | Number of bootstrap iterations to use when computing confidence intervals. |
| units | Identifier of sampling units, which will be used to perform a multilevel bootstrap and account for repeated measures design. |
| orient |**"h" or "v"**  Orientation ko Horizontal ya vertical mein change karny ka liya. |
| color |**Color Name** Graph ka color change karny ka liya |
| palette |**palette name** color combinations   |
| saturation | **int** Color ko halka ya garha karny ky liya. 0.5 for half, 1 for orignal and 2 for double |
| errcolor | Color for all of the error bars. |
| errwidth | Width of the error bars. |
| capsize | **int (0.1 - onwards** Capsize change karny ky liya |
| dodge | If True, the bars are shifted along the categorical axis. |
| ax | Axes object to draw the plot onto, otherwise uses the current Axes. |

## 1.6- Scatterplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head(3)
```

```python
# Syntax
seaborn.scatterplot(data=None, *, x=None, y=None, hue=None, size=None, style=None, palette=None, 
                    hue_order=None, hue_norm=None, sizes=None, size_order=None, size_norm=None, 
                    markers=True, style_order=None, legend='auto', ax=None, **kwargs)
```

```python
# data,x,y

sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip' )

# data = wo data set jis py plot karna hai
# x = x axis py jo column rakhna hai
# y = y axis py jo column rakhna ha

## Yaad rahy dataset ka naam inverted commas kay baghiar rakhna hai( kunka wo defind hai)
## x,y axis py jo column rakhna hai uska naam inverted commas mein rakhna hai(kunka wo defind nhi hen)
```

```python
# hue
sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip',
    hue = 'sex' )

# Agar "hue" mein categorical variable put karen gy to har category kay liya different color show ho ga
# Agar nemeric values put kar rahy hen or unki tadad ziada hai to unky groups ban jaen gy 
```

```python
# palette
sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip',
    hue = 'size',
    palette="deep")

# palette parameter sy colors ko change kar sakty hen "palette" mein color ka naam likhna hai
# colors ky name seaborn color palette py check kar sakty hen.
```

```python
# hue, size, style

sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip',
    hue = 'sex', size = 'size', style = 'time' )

# scatterplot minimum do contineous variables py plot hota hai
# third variable(categorical) "hue" py set kar sakty hen is sy points ka color change ho jay ga.
# fourth variable(categorical) "size" py set kar sakty hen is sy points ka size change ho jay ga.
# fifth variable(categorical) "style" py set kar sakty hen is sy points ka style change ho jay ga.
```

```python
# sizes

sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip',
    hue = 'sex', size = 'size',
    sizes = (20, 200))

# sizes param ko use karty hoy minimum or maximum size ko set kar sakty hen
```

```python
# hue_norm

sns.scatterplot(
    data=df,
    x="total_bill", y="tip",
    hue="size",
    size="size",
    sizes=(20, 200), 
    hue_norm=(0, 7)
)

# hue_norm param ko use karty hoy hue variable ky values ko normalize kar sakty hen
```

```python
sns.scatterplot(
    data=df,
    x="total_bill", y="tip",
    hue="size",
    size="size",
    sizes=(20, 200), 
    hue_norm=(0, 15)
)
```

```python
# marker

sns.scatterplot(
    data = df,
    x ='total_bill', y = 'tip',
    hue = 'sex', size = 'size', style = 'time',
    marker = '+')

# marker param use kar ky marzi ka marker set kar sakty hen
```
